## Introduction:

In this project, I will use supernovae type I-a to estimate the the cosmological parameters for our universe. Namely, I will use data of supernovae with calculated distance moduli to estimate values of $\Omega_{m,0}, \Omega_{\Lambda}$, and $H_0$. This is possible because type Ia supernovae are an example of a standard candle: an astronomical object whose intrinsic luminosity can be calculated. For type Ia supernovae, their peak luminosity can be computed with reasonable accuracy by looking at the shape of its light curve over time. The faster the intensity of the light curve falls off over time, the dimmer it is at its peak luminosity. This means that for a given supernovae, we can measure its light curve to get its absolute magnitude, and measure its flux to compute its apparent magnitude. This gives us enough information to calculate its luminosity distance.

Background: First, I will show how to calculate the luminosity distance for an arbitrary FRW cosmology. The luminosity distance is defined as:

$$    
    d_L^2 = \frac{L}{4 \pi F} 
$$

where L is the object's intrinsic luminosity and F is its measured flux. In an expanding universe, two things occur which cause the luminosity distance to be different from the proper distance. First, as the light travels, it loses energy as it is redshifted by the expanding universe. Second, because the wavelength of light is redshifted, the frequency of detections of a given photon decreases. Both effects mean that the observed flux is scaled by a factor of $a^2$. The result is that the object will look further away than it really is. The luminosity distance is then related to the co-moving distance by

$$
    d_L = r(1+z)
$$

We now need to compute the comoving distance r for a general FRW spacetime (but we will consider the amount of radiation in the universe to be negligible). Starting with the famous FRW metric:

$$
    ds^2 = -c^2dt^2 + a(t)^2\left(\frac{dr^2}{1 - kr^2} + r^2d\theta^2 + r^2\sin{\theta}^2d\phi^2     \right),
$$

The comoving distance will be the time it takes light to travel from us to a given point (or vice versa) along a radial null trajectory. This means we can rewrite the above as 

$$
    \frac{dr}{\sqrt{1-kr^2}} = \frac{c}{a}dt
$$

which for a flat universe (k=0), we find that the comoving distance is given by the integral

$$
    r = \int \frac{c}{a}dt = \frac{c}{H_0}\int_a^1 \frac{da}{a^2\sqrt{\Omega_{m,0}a^{-3} + \Omega_{\Lambda}}}
$$

For a closed universe (positive k), the LHS instead becomes $\frac{1}{\sqrt{k}}\sin^{-1}{(r\sqrt{k})}$, and for an open universe (negative k) it becomes $\frac{1}{\sqrt{k}}\sinh^{-1}{(r\sqrt{k})}$. For non-flat universes, we would also need the curvature term in the integral for a. It will be convenient to make the substiution from a to redshift, $a = \frac{1}{1+z}$:

$$
    r = \frac{c}{H_0}\int_0^z \frac{dz}{\sqrt{\Omega_{m,0}(1+z)^{3} + \Omega_{\Lambda}}}
$$

so putting it all together, we can express the luminosity distance as a function of redshift as 

$$
   d_L(z) =  (1+z)\frac{c}{H_0}\int_0^z \frac{dz}{\sqrt{\Omega_{m,0}(1+z)^{3} + \Omega_{\Lambda}}}
$$

For the actual supernovae data, we will have the distance modulus, which is related to the luminosity distance by

$$
    \mu(z,\Omega_{m,0},\Omega_{\Lambda},H_0) = 5\log_{10}{\left(\frac{d_L(z)}{1 ~\text{Mpc}}\right)} + 25
$$

The procedure then becomes: Plot $\mu$ as a function of $z$. Then, iterate through parameter space for possible values of $H_0, \Omega_{m,0},$ and $\Omega_{\Lambda}$ and quantify the best fit by some statistical metric (such as a chi-square goodness of fit). I also think it would be interesting to do one fit with no $\Lambda$ term, and instead introduce a non-zero curvature to the universe. I expect this fit will be worse, but it will be interesting to see just by how much. Finally, I will do some sort of error analysis to study the covariance between each parameter. The actual data we have is from three seperate datasets, so I will compute the best fit parameters for each dataset seperately, as well as a combination of all them, or perhaps different combinations of two of them.

The calculation of the Hubble constant via this method is relevant for modern day cosmological research. A major problem in the current field of cosmology is the fact that calculting $H_0$ via this method, and from measurements of the CMB power spectrum/ baryon acoustic oscillations in the early univsere (e.g., [Adame et. al 2024](https://ui.adsabs.harvard.edu/abs/2025JCAP...02..021A/abstract)), produce similar but slightly different results, and the difference is greater than what one would expect from measurement uncertainty alone. This phonemoneon is called the Hubble tension (see [Di Valentino et. al (2021)](https://iopscience.iop.org/article/10.1088/1361-6382/ac086d/meta) for a recent review.)


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
import astropy.units as u
import pandas as pd
plt.style.use('mplstyle.txt')
```

All three of the supernovae datasets we will use are from surveys which have made very precise measurements of supernovae absolute magnitudes. Each dataset supplies either an extinction corrected distnce modulus, or an extinction corrected apparent magnitude which we can convert to a distance modulus. First, I will use the supplied dataset from [Betoule et. al (2014)](https://www.aanda.org/articles/aa/full_html/2014/08/aa23413-14/aa23413-14.html#S17). I apply the magnitude correction described in the paper to get better constraints on the distance moduli for supernovae:

$$
    \mu = m_B - M_B - \alpha*X_1 + \beta*C
$$

where $C$ is the color and $X_1$ is a parameter describing the light curve of the supernovae (both are supplied in the dataset.)

```python
z1, mB, emB, x1, C = np.loadtxt('jla_lcparams.txt', usecols=(1,4,5,6,8), unpack=True)

sorted_z1_args = np.argsort(z1)
z1 = z1[sorted_z1_args]
mB = mB[sorted_z1_args]
emB = emB[sorted_z1_args]
x1 = x1[sorted_z1_args]
C = C[sorted_z1_args]

M0sn = -19.05 # assumed universal absolute magnitude
alpha = 0.141
beta = 3.101

mu1 = mB - M0sn + alpha*x1 - beta*C           # distance modulus

z2, mu2, emu2 = np.loadtxt('SCPUnion2.1_mu_vs_z.txt', usecols=(1,2,3), unpack=True)

sorted_z2_args = np.argsort(z2)
z2 = z2[sorted_z2_args]
mu2 = mu2[sorted_z2_args]
emu2 = emu2[sorted_z2_args]



```

Next, I vary $H_0$ and $\Omega_m$ to achieve a best-fit for this dataset using scipy's curve fit function (which does a least squares regression.) Here, I require the universe to be flat, such that $\Omega_{\Lambda} = 1 - \Omega_m$, and is not a free parameter in the model. Later, when I allow for curvature I will allow for all three parameters to vary and see if it converges on a flat universe.

```python
from scipy.optimize import curve_fit

def mu_comb(z, H0, Omega_m):
    Omega_L = 1 - Omega_m
    c = 3e8*u.m/u.s
    H0 = (H0 * u.km/u.s/u.Mpc).to(u.s**-1)


    integrand = lambda z: 1/np.sqrt(Omega_m*(1+z)**3 + Omega_L)
    prefactor = (1+z)*c/H0

    d_L =  (prefactor*quad(integrand, 0, z)[0]).to(u.Mpc)

    mu = 5*np.log10(d_L/(1*u.Mpc)) + 25
    return mu

mu_comb = np.vectorize(mu_comb)

popt1, cov_matrix1 = curve_fit(mu_comb, z1, mu1, sigma=emB, p0=(70, 0.3), bounds=([50, 0], [100, 1]))

print(f"H0 = {popt1[0]}, Omega_m = {popt1[1]}, Omega_L = {1-popt1[1]}")
```

```python
plt.errorbar(z1, mu1, yerr=emB, c='b', fmt='o', ms=0.5, elinewidth=0.1, ecolor="blue")
plt.plot(z1, mu_comb(z1, popt1[0], popt1[1]), c='k', lw=1, label=r"$H_0 = 71.22$ km/s/Mpc, $\Omega_m = 0.27, \Omega_{\Lambda} = 0.73$")
#plt.scatter(z2, mu2, c='r', marker='.', s=2)

plt.xlabel(r'$z$')
plt.ylabel(r'$m-M \; ({\rm mag})$')
plt.xscale('log')
plt.legend(loc='upper left')
plt.show()
```

The best-fit model for this dataset is in good agreement with the current best-fit parameters for $\Omega_m$ and $\Omega_{\Lambda}$ from the Planck 2018 best fit parameters. $H_0$ is obviously a bit higher than the value of the Planck paper, but this is what most people find when determining $H_0$ via this method (i.e., the Hubble tension.)

We can now consider a matter + lambda + curvature universe and see how it compares. For a closed universe with curvature term $\Omega_{k,0} = 1 - \Omega_{m,0} - \Omega_{\Lambda} < 0$, writing the luminosity distance as a function of distance is straightforward given the expressions for comoving distance in curved cosmologies above, and also utilizing the fact that we can express $\Omega_{k,0}$ as $-\frac{k}{H_0^2}$ . With these substitutions, the luminosity distance can be shown to be ([Riess et. al (2004)](https://iopscience.iop.org/article/10.1086/383612/pdf)):

$$
    d_L(z) = \frac{c}{\sqrt{|\Omega_{k,0}|}H_0}(1+z)\sin\left(\sqrt{|\Omega_{k,0}|}\int_0^z \frac{dz}{\sqrt{(1 + \Omega_{m,0}z)(1+z)^2 - z(2+z)\Omega_{\Lambda}}}\right)
$$

For a closed universe. Similarly, for a open universe with $\Omega_{k,0} > 0$, we have

$$
    d_L(z) = \frac{c}{\sqrt{|\Omega_{k,0}|}H_0}(1+z)\sinh\left(\sqrt{|\Omega_{k,0}|}\int_0^z \frac{dz}{\sqrt{(1 + \Omega_{m,0}z)(1+z)^2 - z(2+z)\Omega_{\Lambda}}}\right)
$$

```python
def mu_closed(z, H0, Omega_m, Omega_L):
    c = 3e8*u.m/u.s
    H0 = (H0 * u.km/u.s/u.Mpc).to(u.s**-1)
    Omega_k = abs(1 - Omega_m - Omega_L)

    prefactor = c*(1+z)/(H0*np.sqrt(Omega_k))

    integrand = lambda z: 1/np.sqrt((1 + Omega_m*z)*(1+z)**2 - z*(2+z)*Omega_L)

    d_L = (prefactor*np.sin(np.sqrt(Omega_k)*quad(integrand, 0, z)[0])).to(u.Mpc)

    mu = 5*np.log10(d_L/(1*u.Mpc)) + 25
    return mu

mu_closed = np.vectorize(mu_closed)


def mu_open(z, H0, Omega_m, Omega_L):
    c = 3e8*u.m/u.s
    H0 = (H0 * u.km/u.s/u.Mpc).to(u.s**-1)
    Omega_k = abs(1 - Omega_m - Omega_L)

    prefactor = c*(1+z)/(H0*np.sqrt(Omega_k))

    integrand = lambda z: 1/np.sqrt((1 + Omega_m*z)*(1+z)**2 - z*(2+z)*Omega_L)

    d_L = (prefactor*np.sinh(np.sqrt(Omega_k)*quad(integrand, 0, z)[0])).to(u.Mpc)

    mu = 5*np.log10(d_L/(1*u.Mpc)) + 25
    return mu

mu_open = np.vectorize(mu_open)



    
```

```python
popt_closed1, cov_matrix_closed1 = curve_fit(mu_closed, z1, mu1, sigma=emB, p0=(50, 0.1, 0.1))
popt_closed2, cov_matrix_closed2 = curve_fit(mu_closed, z2, mu2, sigma=emu2, p0=(50, 0.1, 0.1))
popt_closed3, cov_matrix_closed3 = curve_fit(mu_closed, z3, mu3, sigma=emu3, p0=(50, 0.1, 0.1))

print(f"H0 = {popt_closed1[0]}, Omega_m = {popt_closed1[1]}, Omega_L = {popt_closed1[2]}")
print(f"H0 = {popt_closed2[0]}, Omega_m = {popt_closed2[1]}, Omega_L = {popt_closed2[2]}")
print(f"H0 = {popt_closed3[0]}, Omega_m = {popt_closed3[1]}, Omega_L = {popt_closed3[2]}")
```
